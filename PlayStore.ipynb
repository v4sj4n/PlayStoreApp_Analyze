# Google Play Dataset analyze

```python
import numpy as np
import pandas as pd
```

```python
# Run this if you have the full dataset (DONT FORGET TO PUT THE DATASET ON THE SAME FOLDER)
df = pd.read_csv("Google-Playstore.csv")
```

```python
# Run this if you have the sample dataset
df = pd.read_csv("GooglePlayDataSetSample.csv")
```

### 1. Find the app with the highest rating count and the app with the lowest rating count.

```python
df["Released"] = pd.to_datetime(df["Released"])
sorted_by_rating = df.sort_values(by=['Rating Count', 'Minimum Installs', "Released"], ascending=False)

```

```python
print(f"The app wih the least Rating {sorted_by_rating.iloc[-1]['App Name']} \nThe app with the most Rating is {sorted_by_rating.iloc[0]['App Name']}")
```

### 2. Calculate the average rating for each category of apps and identify the category with the highest average rating.


```python
df.groupby('Category').mean()['Rating'].round(decimals=2).sort_values(ascending=False)
```

As we can see the Category with the highest rating is *Role Playing*

### 3.  Sort the dataset by the number of installs in descending order and select the top 10 apps with the most installs.

```python
# Since we have a minimum installs and a maximum number it seems reasonable to calculate an average and use that as the n of installs
df["Average Install Number"] = np.vectorize(lambda x,y : (x + y) /2 )(df["Minimum Installs"], df["Maximum Installs"])
```

```python
df.sort_values("Average Install Number", ascending=False)[0:10]
```

### 4. Calculate the percentage of apps that are free versus paid.


```python
total_n_apps = len(df.index)
n_free_apps = df["Free"].value_counts()[0]
n_paid_apps = df["Free"].value_counts()[1]
```

```python
print(f"The percentage of Free apps is {round(n_free_apps/total_n_apps, ndigits=2)*100}%\nThe percentage of paid apps is {round(n_paid_apps/total_n_apps, ndigits=2)*100}%")
```

### 5. Identify the app with the longest time since its last update and calculate the time difference between the last update and the release date for that app.


```python
df["Last Updated"] = pd.to_datetime(df["Last Updated"])
```

```python
df["Days without update"] = np.where(df["Last Updated"] > df["Released"], (df["Last Updated"] - df["Released"]).dt.days, 0)
```

```python
df.sort_values("Days without update", ascending=False).iloc[0]
```
